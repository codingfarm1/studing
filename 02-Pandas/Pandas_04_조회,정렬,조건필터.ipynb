## **4.2 정렬(sort)**

### **sort_index: index 정렬, 전체 정렬**
- index 기준으로 정렬함 (기본 오름차순 적용)
- 내림차순 정렬 시, ascending=False

```python
df.sort_index().head(5) #오름차순
```

```python
df.sort_index(ascending=False).head(5) #내림차순
```

### **sort_values: 값에 대한 정렬**

- 값의 기준으로 행을 정렬함
- by = '기준이 되는 행'을 설정함
- by에 2개 이상의 컬럼을 지정하여 정렬할 수 있음
- 오름차순/내림차순을 컬럼 별로 지정할 수 있음

```python
df.sort_values(by='age').head() #age를 기준으로 오름차순
```

```python
df.sort_values(by='age', ascending=False).head()  #age를 기준으로 내림차순
```

```python
#문자열 컬럼도 오름차순/내림차순이 알파벳 순서로 정렬됨
df.sort_values(by='class', ascending=False).head()
```

```python
df.sort_values(by=['fare', 'age']).head() #2개 이상의 컬럼을 기준으로 값 정렬 가능
```

```python
df.sort_values(by=['fare', 'age'], ascending=[False, True]).head() #오름차순/내림차순 정렬도 컬럼에 각각 지정 가능
```

## **4.3 Indexing, Slicing, 조건 필터링**

### **loc - indexing / slicing**
- indexing과 slicing 가능
- slicing은 [시작(포함) : 끝(포함)] 규칙

```python
df.loc[5, 'class'] #indexing (5행, class열의 값)
```

```python
df.loc[2:5, ['age', 'fare', 'who']] #fancy indexing (2:5행, age,fare,who열의 값들)
```

```python
df.loc[2:5, 'age':'who'].head() #slicing
```

```python
df.loc[:6, 'class':'deck']
```

### **loc - 조건 필터**

boolean index을 만들어 조건에 맞는 데이터만 추출할 수 있음
- 변수명 = (조건)
- 변수명.loc[변수명]

```python
cond = (df['age'] >= 70) #조건 생성
cond
```

```python
df.loc[cond].head() #조건에 맞는 데이터만 추출
```

### **loc - 다중 조건**

- 다중 조건은 먼저 condition을 정의하고 &와 |연산자로 복합 조건 생성함

```python
# 조건 1 정의
cond1 = (df['fare'] > 30)

# 조건 2 정의
cond2 = (df['who'] == 'woman')

df.loc[cond1 & cond2].head()
```

```python
df.loc[cond1 | cond2].head()
```

```python
cond = (df['age'] >= 70)
df.loc[cond, 'age'] # 조건에 맞는 필터 중 나이 컬럼만 가져오기
```

```python
df.loc[cond, 'age'] = -1 # 조건에 맞는 필터 중 나이 컬럼만 -1로 변경
df.loc[cond]
```

### **isin**

- 특정 값의 포함 여부는 isin 함수를 통해 비교가 가능함

```python
sample = pd.DataFrame({'name': ['kim', 'lee', 'park', 'choi'],
                        'age': [24, 27, 34, 19]})
sample
```

```python
sample['name'].isin(['kim', 'lee'])
```

```python
sample.isin(['kim', 'lee'])
```

loc을 활용한 조건 필터링으로도 찰떡 궁합임

```python
condition = sample['name'].isin(['kim', 'lee'])
sample.loc[condition] #조건에 맞는 데이터만 추출
```
