### **Excel - 저장하기**<br>
DataFrame을 Excel로 저장할 수 있으며, Excel로 저장시 파일명을 지정합니다.
- 변수명 = pd.read_excel('파일명', index = False, sheet_name = , engine = 'openpyxl')

- index=False 옵션은 가급적 꼭 지정하는 옵션입니다. 지정을 안하면 index가 별도의 컬럼으로 저장되게 됩니다.
- index = False : 인덱스가 별도의 컬럼으로 저장되지 않음

- sheet_name을 지정하여, 저장할 시트의 이름을 변경할 수 있습니다.

```python
excel = pd.read_excel('/content/data/서울시대중교통/seoul_transportation.xlsx', sheet_name='철도', engine='openpyxl')
excel.head()
```

```python
# 별도의 시트명 지정 없이 저장
excel.to_excel('sample.xlsx', index=True)
```

```python
# 시트명을 'sample1'로 지정하여 저장
excel.to_excel('sample1.xlsx', index=False, sheet_name='샘플')
```

## **2) CSV (Comma Separated Values)** <br>

 - 한 줄이 한 개의 행에 해당하며, 열 사이에는 쉼표(,)를 넣어 구분합니다.
 - Excel보다는 훨씬 가볍고 차지하는 용량이 적기 때문에 대부분의 파일데이터는 csv 형태로 제공됩니다.

```python
df = pd.read_csv('/content/data/서울시주민등록인구/seoul_population.csv')
df
```

#### **CSV - 저장하기**<br>
저장하는 방법은 excel과 유사합니다. 다만, csv파일 형식에는 sheet_name 옵션은 없습니다

```python
df.to_csv('sample.csv', index=False) #csv 파일형식으로 저장
```

```python
excel = pd.read_excel('data/seoul_transportation.xlsx', sheet_name='버스', engine='openpyxl')
excel.to_csv('sample1.csv', index=False) # excel파일로 읽어와 csv 파일형식으로 저장
```
