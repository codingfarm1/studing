### **6.2.3) 결측치 채우기 - fillna()**
- 결측치에 대하여 **일괄적으로 값을 채울 수 있음**
-- 일정한 값으로 결측치 채우기
-- 평균으로 결측치 채우기
-- 중앙값으로 결측치 채우기
-- 최빈값으로 결측치 채우기

```python
# 다시 원본 DataFrame 로드
df = sns.load_dataset('titanic')

# 원본을 copy하여 df1 변수에
df1 = df.copy()

df1.tail()
```

```python
df1['age'].fillna(700).tail() # 결측치 채우기
```

```python
df1['age'] = df1['age'].fillna(700)
df1.tail()
```

```python
df1['age'].fillna(df1['age'].mean()).tail()  # 평균으로 채우기
```

```python
df1['age'].fillna(df1['age'].median()).tail()  # 중앙값으로 채우기
```

```python
df1['deck'].mode() # 최빈값으로 채우기
df1['deck'].mode()[0] # 최빈값(mode)로 채울 때 0번째 index 지정하여 값을 추출한 후 채워야 함
```

```python
df1['deck'].fillna(df1['deck'].mode()[0]).tail()
```

### **6.2.4) NaN 값이 있는 데이터 제거하기(dropna)**

- dropna()로 1개라도 NaN 값이 있는 행은 제거할 수 있음 (how='any')
- 기본 옵션 값은 how=any로 설정되어 있으며, 다음과 같이 변경이 가능하다
-- any : 1개라도 NaN 값이 존재 시 drop
-- all : 모두 NaN 값이 존재 시 drop

```python
df1.dropna()
df1.dropna(how='all')
```

```python
# 데이터에서 age가 결측치인 승객의 나이를 30세로 일괄 채우기
df.loc[df['age'].isnull(), 'age'] = 30
```

```python
# 여자/남자 승객의 나이의 결측치는 여자/남자 승객의 나이의 평균으로 채우기
female = (df1['sex'] == 'female')
male = (df1['sex'] == 'male')

female_age = df1.loc[female, 'age'].mean()
male_age = df1.loc[male, 'age'].mean()

df1.loc[female, 'age'] = df1.loc[female, 'age'].fillna(female_age)
df1.loc[male, 'age'] = df1.loc[male, 'age'].fillna(male_age)
```

### **pd.cut() - 구간 나누기(binning)**

- 연속된 수치(continuous values)를 **구간으로 나누어 카테고리화 할 때** 사용함
- 직접 범위를 설정할 수 있음
-- right=False 로 지정 시 우측 범위를 포함하지 않음
-- labels를 지정할 수 있음, 지정한 bins의 개수보다 1개 적어야 함

```python
bins = [0, 6000, 100000, df2['이동거리'].max()]
pd.cut(df2['이동거리'], bins, right=False)
```

```python
# labels 지정 가능
labels = ['적음', '보통', '많음']
pd.cut(df2['이동거리'], bins, labels=labels, right=False)
```

```python
# bins : 옵션에 나누고자 하는 구간의 개수를 설정함
df2['이동거리_cut'] = pd.cut(df2['이동거리'], bins=3)
df2['이동거리_cut'].value_counts()
```

위의 분포를 보니 첫 구간에 데이터가 쏠려 있다. <br>
이 이유는 **최소에서 최대 구간을 지정한 bin만큼 동일하게 분할하기 때문**이다.
<br>
고르게 분포한 데이터라면 괜찮지만, 튀는 **이상치(outlier)가 있는 경우 좋지 않음**.

### **pd.qcut() - 동일한 갯수를 갖도록 구간 분할**
- quantity, 즉 데이터의 분포를 최대한 비슷하게 유지하는 구간을 분할함

```python
df2['이동거리_qcut'] = pd.qcut(df2['이동거리'], q=3)
df2['이동거리_qcut'].value_counts()
```

구간도 예쁘게 분할(균등하게 분할)이 된 것 처럼 보이지만, 간격은 일정하지 않습니다.

```python
qcut_bins = [0, 0.2, 0.8, 1]
pd.qcut(df2['이동거리'], qcut_bins)
```

- labels 지정 : 범위 수 보다 1개 적게 지정

```python
qcut_labels = ['적음', '보통', '많음']
pd.qcut(df2['이동거리'], qcut_bins, labels=qcut_labels).value_counts()
```
